```python
import time
import requests # for requesting the browser
import pandas as pd 

from tqdm import tqdm #for checking how long it will take to procces the quote
from bs4 import BeautifulSoup # for webscraping

## WE use these Libraries to scrap the data from Dinamic Sites
import chromedriver_binary
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

link = 'https://stock-pictures.netlify.app'
```

```python
service = Service(chromedriver_binary.chromedriver_filename)
browser = webdriver.Chrome(service = service)
```

```python
browser.get(link)
```

## AUTO INFINITE SCROLL TO REACH THE BOTTOM

```python
for i in tqdm(range(0,1920000,10000)):
    browser.execute_script('window.scrollTo(0,'+str(i)+')')
    time.sleep(1) # time.sleep() is used to repeate the above code again and again till its reaches to the end of the page
```

```python
soup =BeautifulSoup(browser.page_source,'html.parser')
```

## SCRAPING THE DATA 

```python
data = []
for sp in soup.find_all('div', class_ = 'container'):

    if('gif' not in sp.find('img').get('src')):
        link    = sp.find('img').get('src')
    
        Tags    = sp.find('span').text[6:]
    
        Like    = int(sp.find('div', class_ ='likes-comments').find_all('span')[0].text[:-6])
    
        Comment = int(sp.find('div', class_ ='likes-comments').find_all('span')[1].text[:-10])
        #print(link,Tags,Like,Comment)
        data.append([link,''.join(Tags),Like,Comment])
```

## Saving the DATA in a DATASET

```python
df = pd.DataFrame(data , columns = ['Image_link','Tags','Likes','Comments'])
df.head()
```

```python
df = df.drop_duplicates()  
```
